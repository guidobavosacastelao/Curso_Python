

## Optimización 


Definimos una función: $f(x)=x^{2}+10.sin(x)$

```python
def f(x):
  return x**2+10*sin(x)
```

```python
f(3)
```

```python
x=np.linspace(-10,10,51) #creamos 50 puntos para x
x
```

```python
u = np.vectorize(f)
y = u(x)
```

```python
y
```

```python
fig = plt.figure(figsize=(8,6),dpi=90)
plt.plot(x,y)
plt.grid('on')
plt.show()
```

```python
import scipy.optimize as opt
```

Encontrar el minimo de la función f(x) usando el método de BFGS ( Broyden-Fletcher-Goldfarb-Shanno), emprezando desde un punto inicial $x_0=0$

```python
[xopt,fopt,gopt,Bopt,func_calls,grad_calls,warnflg] = \
  opt.fmin_bfgs(u,         #Funcion objetivo a ser minimizada (vectorizada)
              x0=0,        #Punto de partida original
              maxiter=2000, 
              full_output= True)
  
```

```python
xopt
```

```python
fopt
```

```python
fig = plt.figure(figsize=(8,6),dpi=80)
plt.plot(x,y)
plt.plot([0,xopt],[u(0),fopt],'o')
plt.grid('on')
plt.show()
```

Qué pasaría si tuvieramos un punto de partida de $x_0=6$?

```python
[xopt,fopt,gopt,Bopt,func_calls,grad_calls,warnflg] = \
  opt.fmin_bfgs(u,         #Funcion objetivo a ser minimizada (vectorizada)
              x0=6,        #Punto de partida original
              maxiter=2000, 
              full_output= True)
```

```python
xopt
```

```python
fopt
```

```python
fig = plt.figure(figsize=(8,6),dpi=80)
plt.plot(x,y)
plt.plot([6,xopt],[u(6),fopt],'o')
plt.grid('on')
plt.show()
```

Limitaciones del algoritmo: Puede ser atrapado dentro de un mínimo local

```python
res = \
   opt.basinhopping(u,
                    x0=6,
                    T=1, # Parámetro de Temperatura
                    stepsize=2) # Pasos iniciales
res
```

```python
res.x
```

```python
res.fun
```

```python
fig = plt.figure(figsize=(8,6),dpi=80)
plt.plot(x,y)
plt.plot([6,res.x],[u(6),res.fun],'o')
plt.grid('on')
plt.show()
```
