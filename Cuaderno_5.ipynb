El indexador **.loc** selecciona los datos de una manera diferente a como lo hace el operador de indexaci√≥n. Puede seleccionar subconjuntos de filas o columnas. Tambi√©n puede seleccionar simult√°neamente subconjuntos de filas y columnas. Lo m√°s importante, solo selecciona datos por la **ETIQUETA** de las filas y columnas.

````python
df.loc['Niko'] # El indexador .loc devolver√° una sola fila como una Serie cuando se le de una etiqueta de fila √∫nica.
               # Seleccionemos la fila para 'Niko'
````

````python
df.loc[['Niko', 'Penelope']] # Para seleccionar varias filas, colocar todas las etiquetas de fila se deseen seleccionar 
                             # en una lista y pasarlas a .loc. Seleccionemos a Niko y Pen√©lope
````

````python
df.loc['Niko':'Dean']       # Se selecciona un rango de filas con .loc
````

````python
df.loc[:'Aaron']            # Todo hasta Aaron (inclusive)
````

````python
df.loc['Niko':'Christina':2] # Desde Niko hasta Christina en pasos de a dos
````

````python
df.loc['Dean':]              # Desde Dean en adelante y hasta el final
````

Seleccionando filas y columnas simult√°neamente con .loc :

```python
df.loc[row_selection, column_selection]
```




````python
df.loc[['Dean', 'Cornelia'], ['age', 'state', 'score']] # Seleccionamos dos filas y tres columnas
````

Se puede usar cualquier combinaci√≥n de selecciones para filas o columnas para .loc

````python
df.loc[['Dean', 'Aaron'], 'food'] 
````

````python
df.loc['Jane':'Penelope', ['state', 'color']]
````

````python
df.loc[['Jane'],['age']] # Otro ejemplo
````

````python
df.loc['Jane', 'age'] # Continuaci√≥n (join Jane con age)
````

````python
df.loc[:'Dean', 'height':] # Seleccionar todo hasta Dean, y de height en adelante
````

````python
df.loc[:, ['food', 'color']] # Es posible seleccionar todas las filas usando dos puntos.
````

````python
df.loc[['Penelope','Cornelia'], :] # Tambi√©n se puede recurrir a esta notaci√≥n para seleccionar todas las columnas
````

````python
df.loc[['Penelope','Cornelia']]   # Esta l√≠nea de c√≥digo es equivalente a la de arriba. Toma todas las columnas
````

````python
rows = ['Jane', 'Niko', 'Dean', 'Penelope', 'Christina']
cols = ['state', 'age', 'height', 'score']
df.loc[rows, cols]               # Esta es otra manera de asignar los nombres de las filas y columnas
````


El indexador **.iloc** es muy similar a **.loc** pero solo usa ubicaciones enteras para hacer sus selecciones. La palabra **.iloc** en s√≠ misma significa ubicaci√≥n de enteros, por lo que deber√≠a ayudar a recordar lo que hace.

````python
df.iloc[3] # Al pasar un solo entero a .iloc, va a seleccionar una fila como Serie
````

````python
df.iloc[[3]]  # En vez de verlo como una Serie se puede ver como un Dataframe
````

````python
df.iloc[[5, 2, 4]] # Se pueden utilizar una lista de enteros para seleccionar varias filas
                   # No usar df.iloc[5, 2, 4]
````

````python
df.iloc[3:5]      # Toma la posici√≥n 3 hasta la 4 inclusive. A Christina no la incluye
````

````python
df.iloc[3:]     # Desde la fila 3 en adelante
````

````python
df.iloc[3::2] # Desde la fila tres, hasta el final, de a dos
````

````python
print(df['height'].min()) # Minimo
print(df['height'].max()) # Maximo
print(df['height'].count()) # Cantidad
print(df['height'].idxmax()) # El √≠ndice del valor m√°ximo
print(df['height'].idxmin()) # El √≠ndice del valor m√≠nimo
print(df['height'].quantile([.25,.5,.75])) # Los quantiles
print(df['height'].skew()) # Asimetria
print(df['height'].kurtosis()) # Kurtosis
````
